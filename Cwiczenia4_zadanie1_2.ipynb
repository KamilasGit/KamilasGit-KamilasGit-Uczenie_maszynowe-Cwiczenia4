```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/palles77/MachineLearning_Exercise4/main/Employee_Compensation_SF.csv')
```

```python
# function returning column index in a dataframe
def column_index(df, query_cols):
    cols = df.columns.values
    sidx = np.argsort(cols)
    return sidx[np.searchsorted(cols,query_cols,sorter=sidx)]
```

```python
# Health/dental - 19 ///  other benefits - 18 // retirement - 17 // tot.salary - 16
# reading 'Total Benefits' -> column 20, 'Total Compensation' -> column 21
a = 'Health/Dental'
b = 'Other Benefits'
#a = 'Total Benefits'
#b = 'Total Compensation'
#a= 'Total Salary'
#b= 'Retirement'
(column_first, colum_second) = column_index(dataset, [ a, b])
# get sample 1000 rows
dataset_for_clustering = dataset.sample(1000).iloc[:, [column_first, colum_second]].values
```

```python
m = 'complete'
# 'ward', 'complete', 'average', 'single'
```

```python
# Drawing dendrogram
dendrogram = sch.dendrogram(sch.linkage(dataset_for_clustering, method=m))
print(m);
```

```python
f = 'euclidean'
clusterN = 18
model = AgglomerativeClustering(n_clusters = clusterN, affinity=f, linkage=m)
X = dataset_for_clustering
model.fit(X)
labels = model.labels_
```

```python
# Drawing Clusters
for n in range(0, clusterN):
  plt.scatter(X[labels==n, 0], X[labels==n, 1], s=30, marker='o')

plt.title('linkage = '+ m + ', ilość klastrów = ' + str(clusterN) + ', affinity = ' + f)
plt.xlabel(a)
plt.ylabel(b)
plt.show()
```
